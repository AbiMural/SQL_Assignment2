```python
# to use sql and pandas libraries
import sqlite3
import pandas as pd
```

```python
# to read the excel file for analysing the data
df = pd.read_excel('/content/pizza_sales_excel_file.xlsx')
```

```python
# to establish the sql server and creating a memory
conn = sqlite3.connect(':memory:')
```

```python
# to load data frames into sql database
df.to_sql("pizza_sales", conn, index=False, if_exists='replace')
```

```python
# creating a function to read the data frame with sql query
def execute_query(query):
  return pd.read_sql_query(query, conn)
```

```python
# understand the data
execute_query("select * from pizza_sales limit 10")
```

To find Total Revenue of Pizza Sales:
1. Sum of all the prices in the total_price column and named it as Total_Revenue

```python
execute_query("select sum(total_price) as Total_Revenue from pizza_sales")
```

To Find total orders of pizzas:
  1. count of order_id can be used to find total orders of pizzas.
  2. If there exists any duplicate count of order_id, distinct(count) of order_id can be used

```python
# Checking, if there's any duplicate value in order_id
execute_query("""select order_id, count(*)
from pizza_sales
group by order_id
having count(*) > 1;""")
```

```python
# from the results, there exists duplicates in order_id
execute_query("select  count(distinct(order_id)) as Total_orders from pizza_sales;")
```

To find Average Revenue of pizza sales:
1. we can calculate average revenue of pizza sales per year by dividing Total_Revenue by 12
2. Total_revenue = sum of all total_price column in pizza sales
3. we can round off the average_revenue by using round function

```python
execute_query("select round(sum(total_price)/12) as Average_Revenue from pizza_sales;")
```

To find Total number of pizzas sold:

To find Total orders of pizzas on month basis:

1. we have to extract month from order_date column by using %m and assigning the numerals to month name by using case
2. assigning distinct count of order_id as monthly_orders
3. grouping them all by month

```python
execute_query("""select
    case strftime('%m', order_date)
        WHEN '01' THEN 'Jan'
        WHEN '02' THEN 'Feb'
        WHEN '03' THEN 'Mar'
        WHEN '04' THEN 'Apr'
        WHEN '05' THEN 'May'
        WHEN '06' THEN 'Jun'
        WHEN '07' THEN 'Jul'
        WHEN '08' THEN 'Aug'
        WHEN '09' THEN 'Sep'
        WHEN '10' THEN 'Oct'
        WHEN '11' THEN 'Nov'
        WHEN '12' THEN 'Dec'
    end
    as month, count(distinct(order_id)) as monthly_orders
from pizza_sales group by month;
""")
```

To find Total revenue of pizza sales on month basis:

1. we have to extract month from order_date column by using %m and assigning the numerals to month name by using case
2. assigning sum of total_price as total_revenue and rounding off by using round function
3. grouping them all by month

```python
execute_query("""select
    case strftime('%m', order_date)
        WHEN '01' THEN 'Jan'
        WHEN '02' THEN 'Feb'
        WHEN '03' THEN 'Mar'
        WHEN '04' THEN 'Apr'
        WHEN '05' THEN 'May'
        WHEN '06' THEN 'Jun'
        WHEN '07' THEN 'Jul'
        WHEN '08' THEN 'Aug'
        WHEN '09' THEN 'Sep'
        WHEN '10' THEN 'Oct'
        WHEN '11' THEN 'Nov'
        WHEN '12' THEN 'Dec'
    end
    as month, round(sum(total_price)) as total_revenue
from pizza_sales group by month;
""")
```

To find total number of orders placed on day basis:

1. we have to extract day from order_date column by using %w and assigning the numerals to days_of_the_week by using case strftime
2. assigning distinct count of order_id as total_orders
3. grouping them all by days and ordering them by total_orders

```python
execute_query("""select
    order_date,
    case strftime('%w', order_date)
        WHEN '0' THEN 'Sunday'
        WHEN '1' THEN 'Monday'
        WHEN '2' THEN 'Tuesday'
        WHEN '3' THEN 'Wednesday'
        WHEN '4' THEN 'Thursday'
        WHEN '5' THEN 'Friday'
        WHEN '6' THEN 'Saturday'
    end
    as days, count(distinct(order_id)) as total_orders
from
    pizza_sales
    group by days
    order by total_orders desc;""")
```

To find Total orders on time basis:

1. we can use subqueries to find out when will be the most of orders placed.
2. the result is top 10 orders  

```python
execute_query("""select
    order_time,
    total_orders
from
    (select
        order_time,
        count(distinct(order_id)) as total_orders
     from
        pizza_sales
     group by
        order_time) as order_counts
order by
    total_orders DESC
limit 10;""")
```

To find number of pizzas sold by pizza category:
1. using pizza_category,total_pizzas sold and total revenue grouping them by pizza category and ordering them by either total revenue or pizzas sold

```python
execute_query("select pizza_category,sum(quantity) as total_pizzas_sold,sum(total_price) as total_revenue from pizza_sales group by pizza_category order by total_pizzas_sold desc;")
```

To find the total number of orders placed by pizza category in percentage:
1. using common table expression(cte) which saves the table temporarly and by using sub queries we can calcualte the percentage

```python
execute_query("""
with total_orders_cte as (
    select count(order_id) as total_orders
    from pizza_sales
)
select
    pizza_category,
    round(count(order_id) * 100.0 / total_orders_cte.total_orders,2) || '%' as percentage_orders
from
    pizza_sales,
    total_orders_cte
group by
    pizza_category;
""")
```
